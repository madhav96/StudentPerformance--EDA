### Cleaning Data

```python
df.isnull().sum()
```

### No null value. We can drop data that are not useful for our analysis

```python
student = df.drop(['race/ethnicity', 'parental level of education'], axis=1)
student.head()
```

### Relationship Analysis

```python
corelation = student.corr()
```

```python
plt.gcf().set_facecolor('white')
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
sns.pairplot(student)
plt.gcf().set_facecolor('white')
```

```python
sns.relplot(x='math score', y='reading score', hue='gender', data=student)
plt.gcf().set_facecolor('white')
```

```python
sns.relplot(x='math score', y='reading score', hue='lunch', data=student)
plt.gcf().set_facecolor('white')
```

```python
sns.distplot(student['math score'], bins = 5)
plt.gcf().set_facecolor('white')
```

```python

sns.distplot(student['reading score'], bins = 5)
plt.gcf().set_facecolor('white')
```

```python
sns.distplot(student['writing score'], bins = 5)
plt.gcf().set_facecolor('white')
```

```python
sns.catplot(x = 'math score', kind = 'box', data = student)
plt.gcf().set_facecolor('white')
sns.catplot(x = 'reading score', kind = 'box', data = student)
plt.gcf().set_facecolor('white')
sns.catplot(x = 'writing score', kind = 'box', data = student)
plt.gcf().set_facecolor('white')
```
